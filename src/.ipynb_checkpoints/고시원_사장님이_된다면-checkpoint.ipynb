- 2020년 코로나 사태로 인한 재택근무 비율 증가

- 권리금, 보증금, 월세, 대출을 위한 초기 투자 자본 얼마나 필요한지
  - 부동산 사이트 웹크롤링해서 긁어오기

- 손익분기점 대비 원금 회수 기간
  - 선형 원가, 조업도 이익 분석(CVP 분석) 모델에 따르면 (한계비용과 한계수입이 불변함) 손익분기점(판매수량 (X)은 총 소득(TR)과 총 비용(TC)으로 다음과 같이 계산 가능
     
     TR = TC
     P * X = TFC + V * X
     P * X - V * X = TFC
     (P - V) * X = TFC

     X = TFC / (P - V)


  여기서:

  TFC: 총 고정 비용
  P: 단위판매가
  V: 단위변동비


  - 선형 그래프로 나타내기
  - 지역별 도표로 통계내기

- 지역별로 가장 수익성이 좋은 고시원 찾기
