```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rc

import seaborn as sns


# %matplotlib inline
# get_ipython().run_line_magic("matplotlib", "inline")

plt.rcParams["axes.unicode_minus"] = False

rc("font", family="NanumGothic")
```

```python
df = pd.read_csv("../data/gosiwon_계산결과_최종.csv", encoding="utf-8")
```

```python
def draw(index):
    total = df["매물 총액 전처리"][index]
    m_rentm = df["월 임대료 전처리"][index]
    maintenance = df["관리비 전처리"][index]
    m_rentp = df["월순수익_중"][index]

    fff = df["회수기간_하"][index]

    prev_total = total

    
    x=[]
    y=[]

    # month
    #              수정해야할 부분 반복횟수 

    for i in range(120):
        total = total + m_rentm + maintenance - m_rentp

        if total >= 0:
            j = i
            g = m_rentp*i

        #if total <= 0:
            #print(i, ": 순수익", (m_rentp*i-prev_total))

        x.append(m_rentp*i)
        y.append(i)
        

    
    plt.figure(figsize=(10, 8))

    plt.plot(x, y, '-', ms=3, lw=1) #xy데이터 플롯-line

    plt.xlabel('Price')    #x축 이름 
    plt.ylabel('Month')    #y축 이름

    plt.scatter(g, j)

    sns.set_style('whitegrid')
```

```python
draw(1543)
```
